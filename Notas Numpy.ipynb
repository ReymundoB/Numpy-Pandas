**Notas de numpy**

En este cuaderno vamos a contar con las funciones principales que ofrece la libreria para el manejo de datos.


```python
import numpy as np
#Generamos un arreglo
print('Creamos un arreglo: ')
a = np.array([1,2,3,4,5])
print(a)

print('Podemos indicar el tipo de dato almacena un arreglo')
c=np.array([1,2,3,4,5], dtype=str)
print(c)


c=np.array([1,2,3,4,5], dtype=np.complex)
print('Números complejos: ',c )

print('Genera un arreglo con el rango 0 a rango mayor-1')
print(np.arange(0,100))



#Funciones de numpy para arreglos
print('Genera un arreglos con 0: ')
print(np.zeros(10,dtype=int))

print('Gener un arreglo con 1: ')
print(np.ones(10,dtype=int))

print('Generar con arreglo con datos basura:')
print(np.empty(10,dtype=float))

print('Generar con arreglo con datos aleatorios, los primeros 2 numeros son el rango y el 3ro la longitud:')
print(np.random.randint(0,100,10))


#Atributos del arreglo
print('Obtener el valor del arreglo')
print(a[3])

print('Editar un elemento del arreglo')
a[0]=100
print(a)

print('Cantidad de atributos que almacena')
print(a.size)

print('Obtiene la dimension del arreglo')
print(a.ndim)

print('Saber el tipo de dato que almacena')
print(a.dtype)

print('Conocer la dimension y la cantidad de elementos')
print(a.shape)


```

3. SubArreglos

```python
#a[start:end:saltos]

a = np.random.randint(0,100,10)
print('Arreglo: ', a)
print('Sub-arreglo obtenemos del indice 0 al 5: ', a[0:5])
print('Sub-arreglo  del indice 5 al final: ',a[5:])
print('Sub-arreglo del indice 5 al final con saltos de 2: ',a[5::2])

#caso especial donde necesitas generar un listado de indices
print('Sub-arreglo con indices espesificos: ',a[[0,1,3]])

#Podemos estraer info. a partir de un listado de elementos bolean
#debe tener el mismo tamaño del arreglo
print('Con boolean: ', a[[True,False,False,False,False,False,False,False,False,True]])
```

**4. Vectorizar funciones.**

Permite realizar operaciones de una funcion sobre los elementos de un arreglo.

```python
a = np.random.randint(0,100,5)
print('Arreglo original: ',a)

#declaramos la funcion
def operacion(valor):
  return(valor **2)+2

#Si queremos aplicar la funcion a los elementos del arreglo seria mediante
print('Aplicamos la funcion a cada elemento de arreglo y retorna entero')
for x in a:
 print( operacion(x))


#vectorizamos la funcion
vector  = np.vectorize(operacion)

#aplicamos la funcion al arreglo
print('Aplicamos la funcion vectorizada y retorna un arreglo')
vector(a)
```

*5. Copias y vistas.*

Copia: un objeto que se copia del arreglo original. No refleja cambios




Vista: una objeto que apunta al arreglo original y que se reflejan los cambios entre ambos

```python
a = np.random.randint(0,100,5)
b = a.copy()
print('Aplicamos una suma al original, no afecta a la copia pero si a la vista')
a=a+10
print('Arreglo original: ',a)
print('Arreglo copia:',b)

c = a.view()
print('Arreglo tipo vista: ', c)
print('Si modificas la vista tambien afecta al original pero no en la copia')
c[0]=10001
print('Vista',c)
print('Original',a)
print('Copia',b)


#Otra forma de generar vista
d = a[:]
#Saber a que apunta
d.base()
#o
d.base() is a



```

**8. Filtros y condiciones**



```python
#generamos un vector de prueba
a = np.random.randint(0,100,20)
print('Vector: ',a)

#operador condicional
print('Operador condicional')
print('\n', a>70)
print('\n', a[a>70])

#operador logico
print('Operador logico')
print('\n', a[(a>60)&(a<80)])
print('\n',a[(a==100)|(a==90)])

#condicion ALL y ANY
print('All: todos los elementos cumplen con las condiciones, regresa booleano\n')
print(np.all((a>10) & (a<15)))

print('\n Any: al menos 1 elemento cumple \n')
print(np.any((a>10) & (a<15)))


```

**8.1 Condicion Where y Select**

```python
notas = np.random.randint(0,11,10)
print('Calificaciones: \n',notas )

#Where: recibe 3 argumentos,
# 1er la condicion, 2do si cumple, 3ro en caso de no cumplir
#Select:admite 2 argumentos,
# 1ero un arreglo de condiciones, 2do un arreglo opciones
#Evalua la condicion y en caso de cumpli asirga la opcion que le corresponde


print('Funcion Where:\n',np.where(notas >= 7, 'Aprobado','Insuficiente' ))

print('\nFuncion select:\n',)
condicion=[(notas == 10), (notas>=7) & (notas<=9), (notas<=7) ]
opcion=(['Excelente 10'], ['Aprobaste'],['Insuficiente'])
print(np.select(condicion, opcion))


```
